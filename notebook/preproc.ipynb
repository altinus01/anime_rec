# Preprocess data for model

Import packages

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
import ast
```

```python
filtered_data=pd.read_csv('../data/filtered_anime_dataset.csv')
filtered_data.head()
```

### Process 'Genres' column and HOE it

```python
filtered_data['Genres']=filtered_data['Genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x,str) else x)
mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(filtered_data['Genres']),columns=mlb.classes_)
filtered_data_encoded=pd.concat([filtered_data,genre_encoded],axis=1)
filtered_data_encoded.drop('Genres',axis=1,inplace=True)
```

```python
filtered_data_encoded.head()
```

### Export as a csv

```python
filtered_data_encoded.to_csv('../data/ohe_data.csv',index=False)
```
