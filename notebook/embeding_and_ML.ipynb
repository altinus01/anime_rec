```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ohe_data=pd.read_csv('../data/ohe_data.csv')
```

```python
ohe_data
```

```python
def time_filter(df,years_back=10):
    top_time=df['Release year'].max()
    return df[df['Release year']>top_time-years_back]
```

```python
#ohe_data=time_filter(ohe_data,3)
ohe_data
```

```python
vectors=ohe_data.drop(columns=['Name', 'Rating', 'Release year'])
vectors.head()
```

```python
def cosine_similarity_entry(vector,matrix):
    elem=0
    matrix['Similarity']=0
    matrix['Similarity']=matrix['Similarity'].astype(float)
    vector["Similarity"]=0
    vector["Similarity"]=vector["Similarity"].astype(float)
    while elem < matrix.shape[0]:
        matrix.loc[elem, "Similarity"]=cosine_similarity(vector, matrix.loc[[elem]])
        elem+=1
    return matrix
```

```python
mat=cosine_similarity_entry(vectors.loc[[0]],vectors)
mat.sort_values('Similarity',ascending=False,inplace=True)
```

```python
mat
```

```python
new=mat.join(ohe_data[['Name','Image source','Rating','Release year']])
new
```

```python
new=time_filter(new,3)
new
```

```python
def top_n(matrix,n):
    return matrix.iloc(axis=0)[0:n]
```

```python
topn=top_n(new,15)
topn
```

```python
topn=topn.sort_values('Rating',ascending=False)
topn
```
